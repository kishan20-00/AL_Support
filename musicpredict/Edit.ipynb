```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans

def adjust_playlist(df, n_playlists=20):
    # Encode categorical variables
    categorical_cols = ["Emotion", "Time", "Weather", "Language"]
    for col in categorical_cols:
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
    
    # Select relevant features for clustering
    feature_cols = ["Emotion", "Time", "Weather", "Tempo_Min", "Tempo_Max", 
                     "Valence_Min", "Valence_Max", "Energy_Min", "Energy_Max", 
                     "Language", "Playlist size"]
    X = df[feature_cols]

    # Apply K-Means clustering to reassign playlists
    kmeans = KMeans(n_clusters=n_playlists, random_state=42, n_init=10)
    df["Playlist"] = kmeans.fit_predict(X)
    
    return df

# Load CSV
file_path = "music_recommendation_dataset_with_separate_ranges.csv"  # Update with your file path
output_path = "adjusted_music_recommendation.csv"
df = pd.read_csv(file_path)

# Adjust the Playlist column
df_adjusted = adjust_playlist(df, n_playlists=20)

# Save the adjusted dataset
df_adjusted.to_csv(output_path, index=False)

print(f"CSV file with adjusted playlist has been saved as {output_path}")
```
